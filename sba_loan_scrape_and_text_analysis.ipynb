```python
import os
import pandas as pd
import re

# Define the list of files
target_files = [
    'sba_loan_data/sba_disaster_loan_data_fy21.xlsx',
    'sba_loan_data/SBA_Disaster_Loan_Data_FY17.xlsx',
    'sba_loan_data/sba_disaster_loan_data_fy00.xlsx',
    'sba_loan_data/sba_disaster_loan_data_fy20.xlsx',
    'sba_loan_data/SBA_Disaster_Loan_Data_FY11.xlsx',
    'sba_loan_data/sba_disaster_loan_data_fy02.xls',
    'sba_loan_data/sba_disaster_loan_data_fy07.xlsx',
    'sba_loan_data/sba_disaster_loan_data_fy01.xls',
    'sba_loan_data/sba_disaster_loan_data_fy06.xlsx',
    'sba_loan_data/SBA_Disaster_Loan_Data_FY10.xlsx',
    'sba_loan_data/sba_disaster_loan_data_fy05.xlsx',
    'sba_loan_data/SBA_Disaster_Loan_Data_FY13.xlsx',
    'sba_loan_data/SBA_Disaster_Loan_Data_FY09.xlsx',
    'sba_loan_data/SBA_Disaster_Loan_Data_FY08.xlsx',
    'sba_loan_data/SBA_Disaster_Loan_Data_FY12.xlsx',
    'sba_loan_data/sba_disaster_loan_data_fy04.xlsx',
    'sba_loan_data/SBA_Disaster_Loan_Data_FY15.xlsx',
    'sba_loan_data/SBA_Disaster_Loan_Data_FY14.xlsx',
    'sba_loan_data/SBA_Disaster_Loan_Data_FY18.xlsx',
    'sba_loan_data/sba_disaster_loan_data_fy22.xlsx'
]

date_pat = r"fy(\d{2})\.(xls|xlsx)$"

dfs = []

for file in target_files:
    print(f"Checking file: {file}")
    
    if not os.path.exists(file):
        print(f"File not found: {file}")
        continue  
    
    file_name = os.path.basename(file)
    year_match = re.findall(date_pat, file_name, re.I)
    
    if year_match:
        year = year_match[0][0]  
        print(f"Year extracted: {year}")
    else:
        print(f"No fiscal year found for {file_name}. Skipping.")
        continue  
    
    try:
        df = pd.read_excel(file, sheet_name=4, skiprows=4)  
        df['Year'] = year
        dfs.append(df) 
    except Exception as e:
        print(f"Error reading {file_name}: {e}")
        continue  
        
if dfs:
    final_df = pd.concat(dfs, ignore_index=True)
    print("Data concatenated successfully!")
    print(final_df.head())  
else:
    print("No data was processed.")
```

```python
broken_endpoints = [] 
incidents_list = []
for code in flattened_list:
    endpoint = base_url + str(code)
    try:
        response = requests.get(endpoint)
        data = response.json()
        incident_text = data.get("results")[0].get("abstract")
        state_text = data.get("results")[0].get("title")
        incidents_list.append({"disaster_code": code,
                              "description": incident_text,
                              "state": state_text})
    except:
        broken_endpoints.append(endpoint)
```

```python
import regex as re
incident_names = []

for incident in actual_incidents:
    pat1 = re.compile(r"incident: (\D*)\.\sIncident Period:", re.I)
    pat_results = pat1.findall(incident)
    #print(pat_results)

    incident_names.append(pat_results)
incident_names
```
