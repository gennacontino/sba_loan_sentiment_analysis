```python
import pandas as pd
from textblob.classifiers import NaiveBayesClassifier
import nltk
```

```python
sba_loans = pd.read_csv('sba_disasters_train_data.csv')
```

```python
sba_loans.head()
```

```python
len(sba_loans)
```

```python
clean_df = sba_loans.sample(frac=1, random_state=42)
```

```python
train = [ (r['Incident'], r['Climate/N'].strip()) for i, r in clean_df[['Incident', 'Climate/N']].iterrows()] 
```

```python
train
```

```python
cl = NaiveBayesClassifier(train)
```

```python
cl.show_informative_features(20)
```

```python
test_data = pd.read_csv('sba_disasters_test_data.csv')
```

```python
test_data = pd.DataFrame(test_data)[['Incident']]
test_data['classify_result'] = test_data['Incident'].apply(lambda x: cl.classify(x))
```
